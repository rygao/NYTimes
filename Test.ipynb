# Finding NYT article urls

```python
## READ NYTIMES API KEY
with open('api_key.txt') as f:
    api_key = f.read().strip()
```

```python
api_url = "https://api.nytimes.com/svc/search/v2/articlesearch.json"
url_params = urllib.urlencode({
    'api-key': api_key,
    'fq': 'source:("The New York Times") AND section_name:("U.S.") AND document_type:("article")',
    'begin_date': "20151106",
    'end_date': "20151106"
    })
```

```python
# api_data = urllib2.urlopen("%s?%s" % (api_url, url_params)).read()

api_data = requests.get("%s?%s" % (api_url, url_params)).json()
```

# Scraping NYT full articles

```python
## SET UP COOKIES
cj = CookieJar()
opener = urllib2.build_opener(urllib2.HTTPCookieProcessor(cj))
```

```python
%%time

for doc in api_data['response']['docs']:
    url = doc['web_url']
    print url
    
    soup = BeautifulSoup(opener.open(url).read())
    text = soup.title
#     text = ' '.join([e.text for e in soup.findAll("p", "story-body-text story-content")])
    print text
```

```python
def taggify(soup):
    bsTag = type(BeautifulSoup('<a/>').findAll('a')[0])
    for tag in soup:
        if isinstance(tag, bsTag):
            yield '<{}>{}</{}>\n'.format(tag.name,''.join(taggify(tag)),tag.name)
```
